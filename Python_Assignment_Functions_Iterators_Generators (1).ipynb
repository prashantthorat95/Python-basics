### **6. Explain the concept of generators in Python and how they are defined.**
Generators use `yield` instead of `return` to produce a sequence of values lazily.

**Example:**
```python
def gen_numbers():
    for i in range(3):
        yield i

for num in gen_numbers():
    print(num)



```

### **7. . What are the advantages of using generators over regular functions?**
- Memory efficient (no need to store all data)
- Lazy evaluation (generate values on demand)
  Faster Execution for Large Data

Since generators yield results one by one, they avoid the overhead of creating and storing large data structures in memory.

This improves performance for large-scale or streaming data.

 Infinite Sequence Support

Generators can produce infinite sequences (like reading data continuously) because they don’t require all values to be stored at once.

### **9. ` Explain the purpose and usage of the `map()` function in Python.**

The map() function in Python is used to apply a given function to each item in an iterable (like a list, tuple, or set) and return a new iterator containing the results.
It is commonly used for data transformation or element-wise operations.

Syntax:
map(function, iterable)


function → The function to apply to each element.

iterable → The collection of data (list, tuple, etc.).
Applies a given function to all items in an iterable.

**Example:**
```python
nums = [1, 2, 3]
result = list(map(lambda x: x*2, nums))
print(result)
```

**1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.**

````python

def sum_of_even(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Taking list input
numbers = list(map(int, input("Enter numbers separated by spaces: ").split()))

# F call
result = sum_of_even(numbers)
print("Sum of even numbers:", result)
````

**4. Prime Numbers from 1 to 200**

````python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

primes = [x for x in range(1, 201) if is_prime(x)]
print(primes)
````

**5. Fibonacci Iterator Class**

````python
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

for num in Fibonacci(10):
    print(num, end=" ")
````

**6. Generator for Powers of 2**

````python
def power_of_two(n):
    for i in range(n+1):
        yield 2**i

for val in power_of_two(5):
    print(val)
````

**7. Generator to Read File Line by Line**

````python
def read_file(filename):
    with open(filename) as f:
        for line in f:
            yield line.strip()
````

````python
text = "Data Science"
vowels = "aeiouAEIOU"
result = ''.join(filter(lambda ch: ch not in vowels, text))
print(result)
````

**11. Bookshop Accounting using Lambda and Map**

````python
# Given data: [Order number, Book title and author, Quantity, Price per item]
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Using lambda and map
result = list(
    map(
        lambda order: (
            order[0],
            order[2] * order[3] if order[2] * order[3] >= 100 else order[2] * order[3] + 10
        ),
        orders
    )
)

# Display result
print(result)
````
