# **Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

---

## 🔹 Conceptual Explanation

**Anomaly Detection** is a machine learning technique used to identify data points that significantly differ from the majority of the data.

These unusual observations are called:

- Outliers  
- Exceptions  
- Deviations  
- Rare events  

Anomalies often indicate:

- Fraud
- System failure
- Data corruption
- Cyber attacks
- Medical abnormalities

---

## 🔹 Mathematical Explanation

Let dataset:

\[
X = \{x_1, x_2, ..., x_n\}
\]

An anomaly is a point \( x_i \) such that:

\[
P(x_i) \ll \epsilon
\]

Where:
- \( P(x_i) \) = Probability of occurrence
- \( \epsilon \) = Small threshold

In distance-based methods:

\[
d(x_i, \mu) > \tau
\]

Where:
- \( \mu \) = Mean
- \( \tau \) = Threshold
- \( d \) = Distance metric

If distance is very large → anomaly.

---

# 🔹 Types of Anomalies

---

## **1️⃣ Point Anomalies**

### 📌 Definition:
A single data point that is significantly different from others.

### 📌 Mathematical View:
\[
x_i \notin \text{Normal Distribution Range}
\]

### 📌 Example:
- Credit card transaction of ₹5,00,000 when usual spending is ₹5,000
- Temperature reading of 100°C in normal weather data

### 📌 Real-World Use Case:
- Fraud detection
- Intrusion detection
- Manufacturing defect detection

---

## **2️⃣ Contextual Anomalies**

### 📌 Definition:
An observation that is anomalous in a specific context but normal in another.

Context could be:
- Time
- Location
- Season
- Environment

### 📌 Mathematical View:

\[
P(x_i | Context) \ll \epsilon
\]

### 📌 Example:
- 30°C temperature is normal in summer
- 30°C temperature is abnormal in winter

- High website traffic at 8 PM (normal)
- High website traffic at 3 AM (anomaly)

### 📌 Real-World Use Case:
- Time-series analysis
- Weather monitoring
- Seasonal sales tracking

---

## **3️⃣ Collective Anomalies**

### 📌 Definition:
A group of data points that collectively behave abnormally, even if individual points appear normal.

### 📌 Mathematical View:

Let sequence:

\[
S = \{x_1, x_2, ..., x_k\}
\]

Individually normal:

\[
P(x_i) \approx normal
\]

But collectively:

\[
P(S) \ll \epsilon
\]

### 📌 Example:
- Multiple small suspicious login attempts
- Gradual increase in network traffic before cyber attack
- Slight vibration increases in machine before breakdown

### 📌 Real-World Use Case:
- Network intrusion detection
- Sensor monitoring
- Medical ECG signal analysis

---

# 🔹 Intuitive Explanation

Imagine a classroom:

- One student shouting randomly → Point anomaly  
- Student shouting during exam → Contextual anomaly  
- Group of students whispering together → Collective anomaly  

---

# 🔹 Assumptions

- Normal behavior is more frequent than anomalies
- Anomalies are rare
- Features properly represent behavior

---

# 🔹 Advantages

✔ Detects fraud and threats early  
✔ Prevents system failures  
✔ Useful in healthcare and cybersecurity  

---

# 🔹 Disadvantages

❌ Hard to define “normal”  
❌ False positives possible  
❌ Requires domain knowledge  

---

# 🔹 Algorithms Used

- Z-Score
- Isolation Forest
- One-Class SVM
- DBSCAN
- Autoencoders

---

# 🔹 Final Summary

Anomaly Detection identifies rare and unusual patterns in data.

Types:
- Point anomaly → Single abnormal observation  
- Contextual anomaly → Abnormal in specific context  
- Collective anomaly → Group abnormal behavior  

It is widely used in fraud detection, cybersecurity, healthcare, and industrial monitoring.

# **Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**

---

# 🔹 Introduction

Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) are popular algorithms used for **Anomaly Detection**, but they differ in:

- Approach
- Assumptions
- Computational complexity
- Best use cases

---

# 🔹 1️⃣ Isolation Forest

## 📌 Conceptual Approach

Isolation Forest is an **ensemble-based algorithm** that isolates anomalies instead of profiling normal points.

Main idea:
Anomalies are:
- Few
- Different

Therefore:
They are easier to isolate using random splits.

It builds multiple random decision trees.

---

## 📌 Mathematical Idea

Anomaly score:

\[
s(x) = 2^{-\frac{E(h(x))}{c(n)}}
\]

Where:
- \( E(h(x)) \) = Average path length
- \( c(n) \) = Normalization factor
- Shorter path → Higher anomaly score

---

## 📌 Assumptions

- Anomalies are rare
- Anomalies are different

---

## 📌 Advantages

✔ Works well with high-dimensional data  
✔ Fast and scalable  
✔ No distance metric required  

---

## 📌 Disadvantages

❌ May struggle with clustered anomalies  
❌ Randomness can affect stability  

---

## 📌 Suitable Use Cases

- Credit card fraud detection
- Network intrusion detection
- Large-scale anomaly detection
- High-dimensional datasets

---

# 🔹 2️⃣ DBSCAN

## 📌 Conceptual Approach

DBSCAN is a **density-based clustering algorithm**.

It classifies points as:
- Core points
- Border points
- Noise points

Noise points are considered anomalies.

---

## 📌 Mathematical Idea

Neighborhood:

\[
N_\varepsilon(x) = \{ y \mid d(x,y) \leq \varepsilon \}
\]

Core condition:

\[
|N_\varepsilon(x)| \geq MinPts
\]

Points not reachable → Anomalies.

---

## 📌 Assumptions

- Data has dense regions
- Anomalies lie in low-density areas

---

## 📌 Advantages

✔ Detects arbitrary shaped clusters  
✔ Automatically finds number of clusters  
✔ Good for spatial data  

---

## 📌 Disadvantages

❌ Sensitive to ε parameter  
❌ Struggles with varying densities  
❌ Poor performance in high dimensions  

---

## 📌 Suitable Use Cases

- GPS location clustering
- Geospatial anomaly detection
- Outlier detection in 2D/3D data
- Image segmentation

---

# 🔹 3️⃣ Local Outlier Factor (LOF)

## 📌 Conceptual Approach

LOF is a **density-based local anomaly detection algorithm**.

It compares:
Density of a point  
vs  
Density of its neighbors

If density is significantly lower → anomaly.

---

## 📌 Mathematical Idea

Local Outlier Factor:

\[
LOF(x) = \frac{\text{Average Local Reachability Density of Neighbors}}{\text{Local Reachability Density of } x}
\]

If:

\[
LOF(x) \gg 1
\]

→ Anomaly

---

## 📌 Assumptions

- Anomalies have lower density than neighbors
- Local comparison is meaningful

---

## 📌 Advantages

✔ Detects local anomalies  
✔ Works well when density varies  
✔ More precise than global methods  

---

## 📌 Disadvantages

❌ Computationally expensive  
❌ Not scalable to very large datasets  
❌ Sensitive to k parameter  

---

## 📌 Suitable Use Cases

- Financial fraud
- Network intrusion
- Medical abnormality detection
- Moderate-size datasets

---

# 🔹 Comparison Table

| Feature | Isolation Forest | DBSCAN | LOF |
|----------|------------------|--------|------|
| Approach | Tree-based isolation | Density-based clustering | Local density comparison |
| Needs Distance Metric | No | Yes | Yes |
| Handles High Dimensions | Yes | Poor | Moderate |
| Detects Local Anomalies | Moderate | Weak | Strong |
| Scalability | High | Moderate | Low |
| Parameter Sensitivity | Low | High (ε) | High (k) |

---

# 🔹 Intuitive Explanation

Imagine a forest:

- Isolation Forest → Finds trees that are easy to separate  
- DBSCAN → Finds lonely trees far from dense forest  
- LOF → Finds trees that are less dense compared to nearby trees  

---

# 🔹 Final Recommendation Guide

Use **Isolation Forest** when:
- Large dataset
- High-dimensional data
- Need scalability

Use **DBSCAN** when:
- Data is spatial
- Clusters have arbitrary shapes

Use **LOF** when:
- Local anomalies matter
- Density varies across dataset

---

# 🔹 Final Summary

- Isolation Forest isolates anomalies using random trees.
- DBSCAN detects anomalies as noise in density-based clusters.
- LOF detects anomalies by comparing local densities.

Choice depends on:
- Data size
- Dimensionality
- Density structure
- Use case requirements

# **Question 3: What are the key components of a Time Series? Explain each with one example.**

---

# 🔹 Introduction

A **Time Series** is a sequence of data points collected over time at regular intervals.

Examples:
- Daily stock prices
- Monthly sales revenue
- Hourly temperature readings
- Website traffic per minute

A time series is typically composed of four key components:

1. Trend  
2. Seasonality  
3. Cyclical Component  
4. Irregular (Noise) Component  

---

# 🔹 1️⃣ Trend Component

## 📌 Conceptual Explanation

The **Trend** represents the long-term direction of the data.

It shows whether the data is:
- Increasing
- Decreasing
- Constant

It reflects overall growth or decline over time.

---

## 📌 Mathematical Representation

A simple linear trend model:

\[
Y_t = \beta_0 + \beta_1 t
\]

Where:
- \( t \) = Time
- \( \beta_1 \) = Trend slope

If:
- \( \beta_1 > 0 \) → Upward trend
- \( \beta_1 < 0 \) → Downward trend

---

## 📌 Example

- E-commerce sales increasing every year
- Population growth over decades
- Rising global temperatures

---

# 🔹 2️⃣ Seasonality Component

## 📌 Conceptual Explanation

Seasonality represents regular, repeating patterns over fixed time intervals.

These patterns repeat:
- Daily
- Weekly
- Monthly
- Yearly

---

## 📌 Mathematical Representation

Seasonal model:

\[
Y_t = T_t + S_t + E_t
\]

Where:
- \( S_t \) = Seasonal component

---

## 📌 Example

- Ice cream sales increase every summer
- Retail sales spike during Diwali/Christmas
- Higher electricity usage in summer

---

# 🔹 3️⃣ Cyclical Component

## 📌 Conceptual Explanation

Cyclical variation refers to long-term fluctuations that are not fixed or regular like seasonality.

Cycles are:
- Irregular in length
- Often influenced by economic conditions

---

## 📌 Mathematical View

Cyclical component:

\[
Y_t = T_t + C_t + E_t
\]

Where:
- \( C_t \) = Cyclical variation

---

## 📌 Example

- Economic boom and recession cycles
- Real estate market ups and downs
- Stock market bull and bear phases

---

# 🔹 4️⃣ Irregular (Noise) Component

## 📌 Conceptual Explanation

Irregular component consists of random variations that cannot be explained by trend, seasonality, or cycles.

These are:
- Unpredictable
- Short-term fluctuations

---

## 📌 Mathematical Representation

Complete time series model:

\[
Y_t = T_t + S_t + C_t + E_t
\]

Where:
- \( E_t \) = Random error (noise)

---

## 📌 Example

- Sudden stock crash due to news
- Machine breakdown
- Pandemic impact
- Unexpected weather event

---

# 🔹 Additive vs Multiplicative Models

### Additive Model

\[
Y_t = T_t + S_t + C_t + E_t
\]

Used when seasonal variation is constant.

---

### Multiplicative Model

\[
Y_t = T_t \times S_t \times C_t \times E_t
\]

Used when seasonal variation increases with trend.

---

# 🔹 Intuitive Explanation

Imagine a retail shop:

- Trend → Business growing every year  
- Seasonality → More sales during festivals  
- Cyclical → Economic slowdown reduces sales  
- Noise → One-day sales spike due to celebrity visit  

---

# 🔹 Assumptions

- Data collected at regular intervals  
- Components are separable  
- Historical patterns help predict future  

---

# 🔹 Real-World Applications

- Stock market forecasting  
- Weather prediction  
- Demand forecasting  
- Sales prediction  
- Traffic flow prediction  

---

# 🔹 Final Summary

A Time Series consists of:

1. Trend → Long-term movement  
2. Seasonality → Regular repeating patterns  
3. Cyclical → Long-term economic fluctuations  
4. Noise → Random unpredictable variations  

Understanding these components helps in building better forecasting models.

# **Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**

---

# 🔹 1️⃣ What is Stationarity?

## 📌 Conceptual Definition

A time series is said to be **Stationary** if its statistical properties remain constant over time.

These properties include:

- Mean
- Variance
- Autocorrelation

In simple words:
A stationary series does not change its behavior over time.

---

## 🔹 Mathematical Definition

A time series \( Y_t \) is **weakly stationary** if:

1️⃣ Mean is constant:

\[
E(Y_t) = \mu
\]

2️⃣ Variance is constant:

\[
Var(Y_t) = \sigma^2
\]

3️⃣ Covariance depends only on lag (k), not time:

\[
Cov(Y_t, Y_{t-k}) = \gamma(k)
\]

If these conditions hold → Stationary.

---

## 🔹 Types of Stationarity

1. Strict Stationarity  
2. Weak (Covariance) Stationarity  

In practice, we use **Weak Stationarity**.

---

# 🔹 2️⃣ Why is Stationarity Important?

Many forecasting models assume stationarity:

- AR
- MA
- ARMA
- ARIMA

If data is non-stationary:
Model performance becomes unreliable.

---

# 🔹 3️⃣ Intuitive Explanation

Imagine daily temperature:

- If average temperature is stable over years → stationary
- If temperature keeps increasing yearly → non-stationary (trend exists)

---

# 🔹 4️⃣ How to Test Stationarity?

---

## ✅ A) Visual Inspection

Plot time series:

- Upward/downward trend → Non-stationary
- Changing variance → Non-stationary
- Constant mean & variance → Stationary

---

## ✅ B) Rolling Statistics

Compute:
- Rolling mean
- Rolling variance

If they change over time → Non-stationary.

---

## ✅ C) Statistical Tests

### 1️⃣ Augmented Dickey-Fuller (ADF) Test

Null Hypothesis (H₀):
Series has a unit root (Non-stationary)

If:

\[
p < 0.05
\]

Reject H₀ → Stationary

---

### 2️⃣ KPSS Test

Null Hypothesis (H₀):
Series is stationary

If:

\[
p < 0.05
\]

Reject H₀ → Non-stationary

---

# 🔹 5️⃣ How to Transform Non-Stationary Series?

---

## ✅ A) Differencing

Remove trend by subtracting previous value:

\[
Y'_t = Y_t - Y_{t-1}
\]

If trend still exists:
Use second-order differencing.

---

## ✅ B) Log Transformation

Used when variance increases over time:

\[
Y'_t = \log(Y_t)
\]

Stabilizes variance.

---

## ✅ C) Seasonal Differencing

Remove seasonal patterns:

\[
Y'_t = Y_t - Y_{t-m}
\]

Where:
- \( m \) = Seasonal period

Example:
Monthly data → m = 12

---

## ✅ D) Detrending

Fit regression model:

\[
Y_t = \beta_0 + \beta_1 t + \epsilon_t
\]

Subtract trend component.

---

# 🔹 6️⃣ Additive vs Difference Concept

Original Series:

\[
Y_t = T_t + S_t + E_t
\]

After differencing:

\[
Y'_t \approx Stationary
\]

---

# 🔹 7️⃣ Real-World Example

Stock Prices:

- Original prices → Non-stationary
- Daily returns → Stationary

\[
Return_t = Price_t - Price_{t-1}
\]

---

# 🔹 8️⃣ Assumptions

- Data collected at equal time intervals
- Underlying process stable after transformation

---

# 🔹 9️⃣ Advantages of Stationary Series

✔ Easier modeling  
✔ Reliable forecasting  
✔ Better statistical inference  

---

# 🔹 🔟 Disadvantages

❌ Real-world data rarely stationary  
❌ Over-differencing may lose information  

---

# 🔹 Final Summary

A time series is stationary if:

- Mean is constant  
- Variance is constant  
- Covariance depends only on lag  

Testing methods:
- Visual inspection  
- Rolling statistics  
- ADF test  
- KPSS test  

Transformations:
- Differencing  
- Log transformation  
- Seasonal differencing  
- Detrending  

Stationarity is essential for building reliable time-series forecasting models.

# **Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.**

---

# 🔹 Introduction

AR, MA, ARIMA, SARIMA, and SARIMAX are classical **time series forecasting models**.

They are used when:
- Data is sequential
- Time dependency exists
- Forecasting future values is required

They differ in:
- Structure
- Assumptions
- Handling of seasonality
- Use of external variables

---

# 🔹 1️⃣ AR Model (AutoRegressive Model)

## 📌 Structure

AR(p):

\[
Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + ... + \phi_p Y_{t-p} + \epsilon_t
\]

Where:
- \( p \) = number of lag terms
- \( \phi \) = coefficients
- \( \epsilon_t \) = white noise

---

## 📌 Concept

Current value depends on past values.

---

## 📌 Assumptions

- Series must be stationary
- Linear relationship with past values

---

## 📌 Applications

- Stock returns
- Economic indicators
- Short-term forecasting

---

# 🔹 2️⃣ MA Model (Moving Average Model)

## 📌 Structure

MA(q):

\[
Y_t = c + \epsilon_t + \theta_1 \epsilon_{t-1} + ... + \theta_q \epsilon_{t-q}
\]

Where:
- \( q \) = number of past errors
- \( \theta \) = coefficients

---

## 📌 Concept

Current value depends on past forecast errors.

---

## 📌 Assumptions

- Series must be stationary
- Error terms are white noise

---

## 📌 Applications

- Noise smoothing
- Short-term signal modeling

---

# 🔹 3️⃣ ARIMA Model

ARIMA = AutoRegressive Integrated Moving Average

ARIMA(p, d, q)

---

## 📌 Structure

\[
(1 - \sum \phi_i L^i)(1 - L)^d Y_t = (1 + \sum \theta_i L^i)\epsilon_t
\]

Where:
- \( p \) = AR terms
- \( d \) = Differencing order
- \( q \) = MA terms
- \( L \) = Lag operator

---

## 📌 Concept

Combines:
- AR component
- Differencing (to make stationary)
- MA component

---

## 📌 Application

- Non-stationary data
- Sales forecasting
- Demand prediction

---

# 🔹 4️⃣ SARIMA Model

SARIMA = Seasonal ARIMA

SARIMA(p, d, q)(P, D, Q, m)

---

## 📌 Structure

Includes:
- Non-seasonal ARIMA
- Seasonal AR, MA, and differencing

Seasonal component:

\[
(P, D, Q, m)
\]

Where:
- \( m \) = seasonal period

---

## 📌 Concept

Handles both:
- Trend
- Seasonality

---

## 📌 Applications

- Monthly sales
- Weather forecasting
- Electricity demand

---

# 🔹 5️⃣ SARIMAX Model

SARIMAX = SARIMA with eXogenous variables

---

## 📌 Structure

\[
Y_t = SARIMA + \beta X_t
\]

Where:
- \( X_t \) = external variables
- \( \beta \) = coefficient

---

## 📌 Concept

Uses:
- Past values
- Past errors
- Seasonality
- External predictors

---

## 📌 Applications

- Sales forecasting with promotions
- Stock prediction using macroeconomic indicators
- Energy demand using temperature data

---

# 🔹 Comparison Table

| Model | Handles Trend | Handles Seasonality | Handles Exogenous Variables | Requires Stationary |
|--------|--------------|--------------------|-----------------------------|---------------------|
| AR | No | No | No | Yes |
| MA | No | No | No | Yes |
| ARIMA | Yes (via d) | No | No | After differencing |
| SARIMA | Yes | Yes | No | After differencing |
| SARIMAX | Yes | Yes | Yes | After differencing |

---

# 🔹 Intuitive Explanation

Imagine predicting monthly ice cream sales:

- AR → Based on last few months' sales
- MA → Based on past prediction errors
- ARIMA → Handles upward sales trend
- SARIMA → Handles summer season boost
- SARIMAX → Also considers temperature and promotions

---

# 🔹 Assumptions

- Linear relationship
- Errors are white noise
- Data collected at equal intervals

---

# 🔹 Advantages

✔ Strong statistical foundation  
✔ Interpretable parameters  
✔ Good for structured time series  

---

# 🔹 Disadvantages

❌ Requires stationarity  
❌ Sensitive to parameter tuning  
❌ Not ideal for nonlinear patterns  

---

# 🔹 Final Summary

- AR → Past values  
- MA → Past errors  
- ARIMA → Trend + AR + MA  
- SARIMA → Trend + Seasonality  
- SARIMAX → Trend + Seasonality + External variables  

Choice depends on:
- Presence of trend
- Seasonality
- External factors

# **Question 6: Load a time series dataset (AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components**

---

# 🔹 Step 1: Understanding the Dataset

The **AirPassengers dataset** contains:

- Monthly airline passenger numbers
- Time period: 1949–1960
- Clear trend and seasonality

This dataset is commonly used to demonstrate:
- Trend analysis
- Seasonal patterns
- Time series decomposition

---

# 🔹 What is Time Series Decomposition?

Decomposition separates a time series into:

\[
Y_t = T_t + S_t + R_t
\]

Where:

- \( T_t \) → Trend component  
- \( S_t \) → Seasonal component  
- \( R_t \) → Residual (noise) component  

---

# 🔹 Additive vs Multiplicative Model

### Additive Model

\[
Y_t = T_t + S_t + R_t
\]

Used when seasonal variation is constant.

---

### Multiplicative Model

\[
Y_t = T_t \times S_t \times R_t
\]

Used when seasonal variation increases with trend.

AirPassengers dataset usually follows:
👉 Multiplicative model (increasing seasonal variation).

---

# 🔹 What We Will Do

1. Load dataset  
2. Convert date column to datetime  
3. Set index as time  
4. Plot original series  
5. Decompose into:
   - Trend
   - Seasonality
   - Residual

---

# 🔹 Expected Observations

- Upward trend (growth in passengers)
- Strong yearly seasonality
- Random noise component

---

# 🔹 Why Decomposition is Important?

✔ Understand structure  
✔ Build better forecasting models  
✔ Identify seasonality  
✔ Remove trend before modeling  

---

# 🔹 Final Goal

Visually analyze:

- Original time series
- Trend pattern
- Seasonal behavior
- Residual noise

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load dataset
df = pd.read_csv('/content/AirPassengers.csv')

# Convert Month column to datetime
df['Month'] = pd.to_datetime(df['Month'])

# Set Month as index
df.set_index('Month', inplace=True)

# Plot original time series
plt.figure()
plt.plot(df)
plt.title("Original AirPassengers Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

# Decompose time series (Multiplicative model)
decomposition = seasonal_decompose(df['#Passengers'], model='multiplicative', period=12)

# Plot decomposition
decomposition.plot()
plt.show()
```

# **Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.**

---

# 🔹 Objective

We will:

1. Load the AirPassengers dataset  
2. Convert it into proper time series format  
3. Train a SARIMA model  
4. Forecast the next 12 months  
5. Visualize actual + forecast values  

---

# 🔹 Why SARIMA?

SARIMA = Seasonal ARIMA

It handles:

- Trend
- Seasonality
- Autocorrelation

AirPassengers dataset contains:
✔ Upward trend  
✔ Strong yearly seasonality  

Therefore, SARIMA is appropriate.

---

# 🔹 Model Structure

SARIMA(p, d, q)(P, D, Q, m)

Where:

- p → AR order  
- d → Differencing order  
- q → MA order  
- P → Seasonal AR  
- D → Seasonal differencing  
- Q → Seasonal MA  
- m → Seasonal period (12 for monthly data)

---

# 🔹 Mathematical Representation

\[
SARIMA = ARIMA(p,d,q) + Seasonal(P,D,Q,m)
\]

Complete equation:

\[
\Phi_p(L)\Phi_P(L^m)(1-L)^d(1-L^m)^D Y_t =
\Theta_q(L)\Theta_Q(L^m)\epsilon_t
\]

---

# 🔹 Assumptions

- Data is stationary after differencing  
- Errors are white noise  
- Linear dependence  

---

# 🔹 Expected Output

- Blue line → Historical data  
- Orange line → Forecast  
- Confidence interval shaded  

---

# 🔹 Real-World Applications

- Sales forecasting  
- Demand prediction  
- Electricity load forecasting  
- Traffic forecasting  

---

# 🔹 Goal

Predict next 12 months passenger traffic using SARIMA.

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load dataset
df = pd.read_csv('AirPassengers.csv')

# Convert Month column to datetime
df['Month'] = pd.to_datetime(df['Month'])

# Set index
df.set_index('Month', inplace=True)

# Extract passenger data
ts = df['#Passengers']

# Train SARIMA Model
# (p,d,q)(P,D,Q,12)
model = SARIMAX(ts,
                order=(1,1,1),
                seasonal_order=(1,1,1,12),
                enforce_stationarity=False,
                enforce_invertibility=False)

results = model.fit()

# Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# Plot results
plt.figure()
plt.plot(ts)
plt.plot(forecast_mean)
plt.fill_between(conf_int.index,
                 conf_int.iloc[:,0],
                 conf_int.iloc[:,1],
                 alpha=0.3)

plt.title("SARIMA Forecast - Next 12 Months")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.show()

# Print forecast values
print(forecast_mean)
```

# **Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib**

---

# 🔹 Objective

We will:

1. Generate a numerical dataset  
2. Apply Local Outlier Factor (LOF)  
3. Detect anomalies  
4. Visualize results using matplotlib  

---

# 🔹 What is Local Outlier Factor (LOF)?

LOF is a density-based anomaly detection algorithm.

It detects anomalies by comparing:

Density of a data point  
vs  
Density of its neighbors  

If a point has significantly lower density than its neighbors → anomaly.

---

# 🔹 Mathematical Insight

Local Reachability Density (LRD):

\[
LRD(x) = \frac{1}{\text{Average Reachability Distance}}
\]

LOF Score:

\[
LOF(x) = \frac{\text{Average LRD of Neighbors}}{LRD(x)}
\]

If:

\[
LOF(x) \gg 1
\]

→ Anomaly

---

# 🔹 Key Parameter

- n_neighbors → Number of nearest neighbors
  - Smaller value → Sensitive detection
  - Larger value → Smoother detection

---

# 🔹 Assumptions

- Normal points have similar density
- Anomalies lie in low-density regions
- Distance metric meaningful

---

# 🔹 Expected Output

- Blue points → Normal observations  
- Red points → Anomalies  

---

# 🔹 Real-World Applications

- Fraud detection  
- Network intrusion detection  
- Medical abnormality detection  
- Manufacturing defect detection  

---

# 🔹 Goal

Detect density-based anomalies using LOF and visualize them in 2D.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs

# Generate synthetic dataset
X, _ = make_blobs(n_samples=300, centers=1, cluster_std=1.0, random_state=42)

# Add some artificial anomalies
rng = np.random.RandomState(42)
X_outliers = rng.uniform(low=-6, high=6, size=(20, 2))
X = np.vstack([X, X_outliers])

# Apply LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)
y_pred = lof.fit_predict(X)

# Separate normal and anomalies
normal = X[y_pred == 1]
anomaly = X[y_pred == -1]

# Plot results
plt.figure()
plt.scatter(normal[:, 0], normal[:, 1])
plt.scatter(anomaly[:, 0], anomaly[:, 1])
plt.title("Local Outlier Factor (LOF) - Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Print number of anomalies detected
print("Number of anomalies detected:", len(anomaly))
```
