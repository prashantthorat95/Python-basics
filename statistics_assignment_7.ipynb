**Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

Descriptive Statistics and Inferential Statistics are the two main branches of statistics, each serving a different purpose in data analysis.

**1. Descriptive Statistics**

Descriptive statistics are used to summarize and describe the main features of a dataset. They help to present data in a meaningful way, making it easier to understand patterns and trends.

It deals with organizing, displaying, and describing data using numerical measures and graphical methods.

Common tools: Mean, Median, Mode, Range, Variance, Standard Deviation, Charts, and Graphs.

**Example:**

Suppose you have the exam scores of 100 students in a class.
You can use descriptive statistics to summarize the data:

Mean score = 75

Median score = 78

Standard deviation = 5

A bar chart showing score distribution

This helps to describe what the data looks like, but not to make predictions beyond this group.

---


**2. Inferential Statistics**

Inferential statistics are used to make predictions, decisions, or generalizations about a population based on a sample of data.
It involves using probability theory to infer conclusions that go beyond the immediate data.

It helps answer questions like: “What can we say about the population from this sample?”

Common tools: Hypothesis Testing, Confidence Intervals, Regression Analysis, ANOVA

 **Example:**

Suppose you take a random sample of 100 students from a university of 10,000 students and find their average height to be 170 cm.
You can use inferential statistics to estimate the average height of all students in the university (the population) — e.g.,

“We are 95% confident that the average height of all students is between 168 cm and 172 cm.”

---

**Summary Table**
| Aspect         | Descriptive Statistics               | Inferential Statistics                                   |
| -------------- | ------------------------------------ | -------------------------------------------------------- |
| **Purpose**    | To summarize and describe data       | To make inferences or predictions about a population     |
| **Data Used**  | Whole dataset (population or sample) | Sample data only                                         |
| **Techniques** | Mean, Median, Mode, Graphs, Charts   | Hypothesis Testing, Confidence Intervals, Regression     |
| **Example**    | Average marks of 100 students        | Predicting average marks of all students in a university |

**Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.**

**Sampling** is the process of selecting a small group (sample) from a larger population to collect data and draw conclusions about the entire population.
Since it is often impractical, time-consuming, or expensive to gather data from every individual, sampling helps researchers make reliable estimates or inferences using only a portion of the population.




**Example:**

If a company has 10,000 employees and wants to know the average satisfaction level, it may survey only 500 randomly selected employees. The results from this smaller group (sample) are then used to estimate satisfaction levels of all employees (population).


---

**1. Random Sampling**

In random sampling, every member of the population has an equal chance of being selected.
This ensures that the sample is unbiased and represents the population fairly.

**Example:**

If you have 10,000 students in a university, and you randomly pick 200 students using a random number generator, each student has an equal probability of being chosen.

**Advantages:**

Simple and easy to perform.

Minimizes selection bias.

**Disadvantages:**

May not represent all subgroups if the population is diverse.


---


**2. Stratified Sampling**

In stratified sampling, the population is first divided into subgroups (strata) based on certain characteristics such as gender, income, age, or region.
Then, random samples are taken from each subgroup proportionally to their size in the population.

**Example:**

In a university where 60% of students are male and 40% are female, researchers may divide the population into two groups (males and females) and then select 120 males and 80 females randomly to form a 200-student sample.
This ensures both genders are represented according to their proportions.

**Advantages:**

Produces more accurate and representative results.

Reduces sampling error, especially when subgroups differ significantly.

**Disadvantages:**

More complex to design and requires prior knowledge of population structure.



---
**Summary Table**

| Feature            | Random Sampling                                | Stratified Sampling                                                        |
| ------------------ | ---------------------------------------------- | -------------------------------------------------------------------------- |
| **Definition**     | Every member has an equal chance of selection. | Population divided into subgroups, and random samples are taken from each. |
| **Representation** | May not ensure representation of all groups.   | Ensures representation of all subgroups.                                   |
| **Complexity**     | Simple and straightforward.                    | More complex and structured.                                               |
| **Example**        | Randomly selecting 200 students from 10,000.   | Selecting 120 males and 80 females to match population proportions.        |
| **Best For**       | Homogeneous populations.                       | Heterogeneous populations.                                                 |



**Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.**



**Answer:**

**Mean, Median, and Mode** are the three most commonly used **measures of central tendency** in statistics.  
They help summarize a large set of data into a single representative value, making it easier to understand and compare datasets.

---

### **1. Mean (Average)**

The **mean** is the **sum of all observations divided by the total number of observations**.  
It is the most widely used measure of central tendency and gives the arithmetic average of a dataset.

**Formula:**  

Mean = (Sum of all values) ÷ (Number of values)


**Example:**  
If the ages of five employees are 22, 25, 28, 30, and 35 years:  

mean = (22 + 25 + 28 + 30 + 35)/5 = 140/5 = 28

So, the **mean age** is **28 years**.

**Advantages:**
- Easy to calculate and understand.  
- Uses all data points, giving a complete picture of the dataset.

**Disadvantages:**
- Sensitive to **extreme values (outliers)** — for example, if one person’s age were 80, the mean would rise significantly.

---

### **2. Median**

The **median** is the **middle value** of a dataset when all observations are arranged in ascending or descending order.  
If there is an even number of observations, the median is the **average of the two middle values**.

**Example:**  
For the same data — 22, 25, 28, 30, 35  
The **median** is the middle value → **28**.  

If the data were 22, 25, 28, 30, 35, 40 (even number of observations):  

Median = (28 + 30)/2 = 29


**Advantages:**
- Not affected by extreme values.  
- Useful for skewed distributions.

**Disadvantages:**
- Does not use all data points.  
- Less sensitive to changes in values.

---

### **3. Mode**

The **mode** is the **value that occurs most frequently** in a dataset.  
A dataset may have **one mode (unimodal)**, **more than one mode (bimodal/multimodal)**, or **no mode** if all values occur equally often.

**Example:**  
In the dataset 4, 5, 6, 6, 7, 8, 8, 8, 9  
The **mode** is **8**, since it appears most frequently.

**Advantages:**
- Can be used for both numerical and categorical data.  
- Easy to identify from data frequency.

**Disadvantages:**
- Not always unique (can have multiple modes).  
- Not useful for continuous data with many unique values.

---

### **Importance of Mean, Median, and Mode**

These measures of central tendency are important because they:

1. **Summarize large data** into a single representative value.  
2. **Identify the center point** of the data distribution.  
3. **Help compare datasets** (e.g., comparing average income or median age between cities).  
4. **Guide decision-making** — businesses, researchers, and policymakers use them to understand trends and patterns.

**Example in real life:**
- **Mean:** Average marks of students in a class.  
- **Median:** Middle income of a population (used to avoid distortion from very high or low incomes).  
- **Mode:** Most common shoe size sold in a store.

---

### **Summary Table**

| Measure | Definition | Example | Key Point |
|----------|-------------|----------|------------|
| **Mean** | Sum of all values ÷ Number of values | (22 + 25 + 28 + 30 + 35) / 5 = 28 | Sensitive to outliers |
| **Median** | Middle value when data is ordered | 22, 25, **28**, 30, 35 | Not affected by outliers |
| **Mode** | Most frequent value | 4, 5, 6, 6, 7 → **6** | Useful for categorical data |

### **Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**

**Answer:**

---

### **1. Skewness**

**Skewness** is a measure of the **asymmetry** or **lack of symmetry** in the distribution of data.  
It tells us whether the data values are spread out more to the left or the right of the mean.

- When data is **perfectly symmetrical**, the mean, median, and mode are all equal, and skewness = 0.
- When data is **asymmetrical**, it is said to be **skewed** — either positively or negatively.

---

#### **Types of Skewness**

1. **Positive Skew (Right-Skewed Distribution):**
   - The **tail of the distribution extends to the right** (toward higher values).
   - Most data points are concentrated on the left side.
   - **Mean > Median > Mode**
   
   **Example:**  
   Income distribution — most people earn moderate salaries, but a few have very high incomes pulling the mean to the right.

2. **Negative Skew (Left-Skewed Distribution):**
   - The **tail of the distribution extends to the left** (toward lower values).
   - Most data points are concentrated on the right side.
   - **Mean < Median < Mode**
   
   **Example:**  
   Age at retirement — most people retire around 60–65, but a few retire much earlier, pulling the mean to the left.

---

### **What Does a Positive Skew Imply About the Data?**

A **positive skew** means that:
- The **majority of the data values are smaller than the mean**.  
- There are **a few unusually large values** (outliers) pulling the distribution to the right.  
- The **mean** is greater than the **median**.  

In simple terms, **most observations are on the lower side**, with a **long right tail**.

**Example:**  
If most people score between 40–70 on a test, but a few students score above 90, the distribution is **positively skewed**.

---

### **2. Kurtosis**

**Kurtosis** measures the **peakedness or flatness** of a data distribution compared to a normal (bell-shaped) curve.  
It describes how heavy or light the tails of a distribution are.

#### **Types of Kurtosis:**

1. **Mesokurtic:**  
   - Normal distribution (standard bell curve).  
   - **Kurtosis ≈ 3** (or 0 if using excess kurtosis).  
   - Moderate tails.

2. **Leptokurtic:**  
   - **Kurtosis > 3**  
   - Distribution is **more peaked** with **heavier tails** (more outliers).  
   - Indicates data values are more concentrated near the mean but with extreme outliers.

3. **Platykurtic:**  
   - **Kurtosis < 3**  
   - Distribution is **flatter** than normal with **lighter tails**.  
   - Indicates less variability and fewer extreme values.

---

### **Summary Table**

| Measure | Description | Type | Shape | Example |
|----------|-------------|------|--------|----------|
| **Skewness** | Measures asymmetry of data | Positive Skew | Tail extends right | Income distribution |
|              |                             | Negative Skew | Tail extends left  | Age at retirement |
| **Kurtosis** | Measures peakedness of data | Leptokurtic | Tall, thin peak (heavy tails) | Stock returns |
|              |                             | Mesokurtic | Normal curve | Standard normal data |
|              |                             | Platykurtic | Flat, light tails | Uniform distribution |

---

**In summary:**  
- **Skewness** tells us about the **direction of asymmetry** in the data.  
- **Kurtosis** tells us about the **sharpness or flatness** of the curve.  
- A **positive skew** means **most values are low**, but a few **high outliers** stretch the distribution’s right tail.


- **Skewness:** Measures the asymmetry of the distribution of data. A symmetric distribution has skewness = 0.
- **Kurtosis:** Measures the 'tailedness' or peak of the data distribution.

A **positive skew** means the right tail of the distribution is longer. This implies that most data values are concentrated on the left side, with a few larger values stretching the mean to the right.

**Question 5: Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]**

````python

# Import statistics module
import statistics as stats

# Given data
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Compute mean, median, and mode
mean_value = stats.mean(numbers)
median_value = stats.median(numbers)
mode_value = stats.mode(numbers)

# Display results
print("Given Numbers:", numbers)
print(f"Mean   = {mean_value}")
print(f"Median = {median_value}")
print(f"Mode   = {mode_value}")
````

### **Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python.**

**Given Data:**
```python
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]


````python

import numpy as np

# Given data
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculate covariance matrix
cov_matrix = np.cov(list_x, list_y, bias=False)
covariance = cov_matrix[0, 1]

# Calculate correlation coefficient
correlation = np.corrcoef(list_x, list_y)[0, 1]

# Display results
print("List X:", list_x)
print("List Y:", list_y)
print("\nCovariance =", covariance)
print("Correlation Coefficient =", correlation)
````

### **Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result.**

**Given Data:**
```python
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


````python

import numpy as np
import matplotlib.pyplot as plt

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Calculate quartiles
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Determine outlier boundaries
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_limit or x > upper_limit]

# Print summary
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Limit: {lower_limit}")
print(f"Upper Limit: {upper_limit}")
print(f"Outliers: {outliers}")

# Plot boxplot
plt.figure(figsize=(6, 5))
plt.boxplot(data, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot of Given Data")
plt.ylabel("Values")
plt.grid(True)
plt.show()
````

### **Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.**

---

#### **Part 1: Conceptual Explanation**

The goal is to understand whether an increase in **advertising spend** leads to an increase in **daily sales** — in other words, if the two variables are related.

To explore this relationship, we use **Covariance** and **Correlation**.

---

### **1. Covariance**

Covariance measures the **direction of the relationship** between two variables:
- A **positive covariance** means that when advertising spend increases, daily sales also tend to increase.
- A **negative covariance** means that as one variable increases, the other tends to decrease.
- A covariance near **zero** indicates no linear relationship.

**Formula:**  
Cov(X, Y) = Σ[(Xi - X̄) × (Yi - Ȳ)] / (n - 1)

However, covariance alone does **not tell us the strength** of the relationship — only the direction.

---

### **2. Correlation**

Correlation measures both the **direction and strength** of the relationship between two variables.  
It is a **standardized version** of covariance that always lies between **-1 and +1**.

- **+1** → Perfect positive correlation (as advertising increases, sales increase proportionally).  
- **-1** → Perfect negative correlation (as advertising increases, sales decrease).  
- **0** → No linear relationship.

**Formula:**  
r = Cov(X, Y) / (σx × σy)

---

### **Interpretation for the Marketing Team:**
By calculating the **correlation coefficient**, we can determine:
- Whether advertising and sales move together.
- How strongly they are related.
- Whether increasing ad spend could potentially lead to higher sales.

---

#### **Part 2: Python Implementation**
We will calculate the **correlation coefficient** between advertising spend and daily sales using Python’s **NumPy** library.


````python
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute covariance matrix
cov_matrix = np.cov(advertising_spend, daily_sales)
covariance = cov_matrix[0, 1]

# Compute correlation coefficient
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

# Display results
print("Advertising Spend:", advertising_spend)
print("Daily Sales:", daily_sales)
print("\nCovariance =", covariance)
print("Correlation Coefficient =", correlation)
````

### **Interpretation of Results**

- The **positive covariance (84,875)** indicates that both variables increase together — as **advertising spend rises, daily sales also increase**.  
- The **correlation coefficient (~0.993)** shows a **very strong positive linear relationship** between advertising and sales.  
- Therefore, investing more in advertising is **strongly associated with higher daily sales**.

---

### **Conclusion**

From this analysis, we can confidently inform the marketing team that **advertising spend and daily sales have a strong positive correlation**, suggesting that **higher advertising investment tends to drive higher sales**.


### **Question 9: Customer Satisfaction Data Analysis**

Your team has collected customer satisfaction survey data on a scale of **1–10** and wants to understand its **distribution** before launching a new product.

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

---

### **Part 1: Explanation**

To understand how customers rated their satisfaction, we can use **summary statistics** and **visualizations**.

---

#### **1. Summary Statistics**

These help describe the **center, spread, and shape** of the data.

- **Mean (Average):**  
  Shows the overall average satisfaction level.  
  A higher mean (closer to 10) indicates higher satisfaction.

- **Median:**  
  The middle value when data is ordered — useful if the distribution is skewed.

- **Mode:**  
  The most frequently occurring score — shows what most customers rated.

- **Standard Deviation (SD):**  
  Measures how spread out the scores are from the mean.  
  A **low SD** means most ratings are similar; a **high SD** means opinions vary widely.

---

#### **2. Visualizations**

Visualizations help us quickly see **patterns, trends, and outliers**.

- **Histogram:**  
  Displays the **frequency distribution** of satisfaction scores.  
  It helps identify whether the data is:
  - **Left-skewed** (most customers gave high scores),
  - **Right-skewed** (most customers gave low scores), or
  - **Normal** (symmetrical distribution around the mean).

- **Boxplot (optional):**  
  Shows median, quartiles, and outliers for deeper insight.

---

#### **Interpretation Goal:**
By using mean, median, standard deviation, and a histogram, we can answer:
- Are customers generally satisfied (high mean)?
- Is satisfaction consistent (low SD)?
- What’s the most common rating (mode)?
- Are there outliers or polarized opinions?

---

### **Part 2: Python Implementation**
We’ll calculate the summary statistics and plot a histogram of the survey scores.


````python
import numpy as np
import matplotlib.pyplot as plt
import statistics as stats

# Given survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Calculate summary statistics
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)
mode_score = stats.mode(survey_scores)
std_dev = np.std(survey_scores, ddof=1)  # Sample standard deviation

# Print summary
print("Survey Scores:", survey_scores)
print(f"Mean Score   = {mean_score:.2f}")
print(f"Median Score = {median_score}")
print(f"Mode Score   = {mode_score}")
print(f"Standard Deviation = {std_dev:.2f}")

# Plot histogram
plt.figure(figsize=(7,5))
plt.hist(survey_scores, bins=6, color='lightblue', edgecolor='black')
plt.title("Customer Satisfaction Survey Scores")
plt.xlabel("Satisfaction Score (1–10)")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
````

### **Interpretation**

- **Mean (7.33)** and **Median (7)** are high → overall customer satisfaction is **good**.  
- **Mode (7)** shows that most customers rated around **7**.  
- **Standard Deviation (1.63)** indicates **moderate variation**, meaning most customers have similar opinions.  
- The **histogram** will likely show a **slight left skew**, suggesting that the majority of customers gave **high satisfaction ratings**.

---

###  **Conclusion**

Customers show **strong positive satisfaction** with only small variations — an encouraging sign before launching the new product.
