### **Question 1: What is hypothesis testing in statistics?**

**Answer:**



### **Simple Explanation**

**Hypothesis testing** is a way to make decisions using data.

It helps us check if a claim (statement) about something is **true or not**, using a **sample**.



### **Think of it like this:**

You hear a claim:

- *"The average height of students in this class is 160 cm."*

You collect some sample data (measure a few students), and then you check:

 *Does the sample data support this claim or not?*



### **What we do in hypothesis testing**

1. **Assume the claim is true** (this is called the **null hypothesis**, H‚ÇÄ).  
2. **Look at the sample data**.  
3. **Check how likely that data is**, if the claim were actually true.  
4. If the data looks **too unlikely**, we say:  
   üëâ ‚ÄúThis claim is probably not true.‚Äù  
   and we **reject the null hypothesis**.  

If the data looks **normal and expected**, we say:  
 ‚ÄúThere is no reason to reject the claim.‚Äù  



### **A Simple Real-Life Example**

Your friend says:

> ‚ÄúMy coin is fair.‚Äù

So H‚ÇÄ: The coin is fair.

You flip it 20 times and get **18 heads**.

This result is very unlikely for a fair coin, so you reject the claim.



### **In short:**

Hypothesis testing helps us decide:

 *Does the sample support the claim or not?*  
 If the sample agrees ‚Üí keep the claim  
 If the sample strongly disagrees ‚Üí reject the claim  



**Very simple definition:**  
 Hypothesis testing is a way to check if what someone claims is likely true, using sample data.**

### **Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer:**

---



The **null hypothesis** is the **default assumption**.

It says that:

- Nothing special is happening  
- There is **no difference**  
- There is **no change**  
- There is **no effect**

In simple words:  
**‚ÄúEverything is normal.‚Äù**



 Alternative Hypothesis (H‚ÇÅ or H‚Çê)

The **alternative hypothesis** is the **opposite** of the null hypothesis.

It says that:

- Something **is happening**  
- There **is a difference**  
- There **is a change**  
- There **is an effect**

In simple words:  
**‚ÄúSomething new or different is happening.‚Äù**



### **3. Simple Example**

A company claims the average battery life of a phone is **10 hours**.

- **H‚ÇÄ (Null):** Average battery life = 10 hours  
  (‚ÄúThe company claim is true.‚Äù)

- **H‚ÇÅ (Alternative):** Average battery life ‚â† 10 hours  
  (‚ÄúThe claim is false; actual battery life is different.‚Äù)

You collect sample data and check which hypothesis the data supports.

---

### **4. Key Differences in Simple Words**

| Null Hypothesis (H‚ÇÄ) | Alternative Hypothesis (H‚ÇÅ) |
|----------------------|-----------------------------|
| Nothing changed | Something changed |
| No effect | Effect exists |
| No difference | Difference exists |
| Default assumption | What you want to test |
| Assume true at the start | Try to find evidence to support it |

---




**In short:**  
H‚ÇÄ says ‚Äúnothing is happening.‚Äù
H‚ÇÅ says ‚Äúsomething is happening.‚Äù

You use sample data to decide which statement is more believable.

### **Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer:**

---

### **1. What is the Significance Level (Œ±)?**

The **significance level**, written as **Œ±**, tells us:

**How much risk we are willing to take of being wrong when we reject the null hypothesis (H‚ÇÄ).**

It is usually set to:

- **0.05 (5%)**  
- **0.01 (1%)**  
- **0.10 (10%)**

Example:  
If Œ± = 0.05, it means we accept a **5% chance** of wrongly rejecting H‚ÇÄ.



### **2. Why do we need a significance level?**

Because when we make decisions using data, there is always some **uncertainty**.  
Œ± helps us control **how strict or lenient** we want our decision to be.



### **3. How does Œ± affect the decision?**

We compare the **p-value** with **Œ±**:

- If **p-value < Œ±** ‚Üí **Reject H‚ÇÄ**  
  (‚ÄúThere is strong evidence against the null hypothesis.‚Äù)

- If **p-value ‚â• Œ±** ‚Üí **Fail to reject H‚ÇÄ**  
  (‚ÄúThere is not enough evidence; keep the null hypothesis.‚Äù)



### **4. Simple Example**

A company says the average weight of their chips packet is **50g**.

You test a sample and get a **p-value of 0.03**.

If Œ± = 0.05:

- 0.03 < 0.05 ‚Üí **Reject H‚ÇÄ**  
  Meaning: The packet weight is likely **not** 50g.

If Œ± = 0.01:

- 0.03 > 0.01 ‚Üí **Fail to reject H‚ÇÄ**  
  Meaning: Not enough evidence to say the weight is wrong.

Same data ‚Üí different decisions depending on **Œ±**.



### **5. Easy Way to Remember**

- **Lower Œ± ‚Üí Stricter testing** (need stronger evidence to reject H‚ÇÄ)  
- **Higher Œ± ‚Üí More tolerant** (easier to reject H‚ÇÄ)



### **In simple words:**

**Significance level tells us how sure we want to be before rejecting the claim (H‚ÇÄ).**  
It is the **cutoff** for deciding whether the result is strong enough to reject H‚ÇÄ.

### **Question 4: What are Type I and Type II errors? Give examples of each.**

**Answer:**

---

### **Type I Error (False Positive)**

A **Type I error** happens when we:

**Reject the null hypothesis (H‚ÇÄ) even though it is actually true.**

In simple words:  
**You think something is happening, but it‚Äôs not.**



### **Example of Type I Error**

- A medical test says a **healthy person has a disease**.  
- A company rejects a **good-quality** product thinking it is defective.  
- A court **punishes an innocent person**.

This error is controlled by the **significance level (Œ±)**.



###  **Type II Error (False Negative)**

A **Type II error** happens when we:

 **Fail to reject the null hypothesis (H‚ÇÄ) even though it is false.**

In simple words:  
**You fail to see something that really exists.**



### **Example of Type II Error**

- A medical test says a **sick person is healthy**.  
- A company accepts a **defective product** thinking it is good.  
- A court **lets a guilty person go free**.

This error is related to **Œ≤ (beta)** and test **power**.

---

###  **Simple Comparison Table**

| Error Type | Meaning | Simple Meaning | Example |
|-----------|---------|----------------|---------|
| **Type I Error** | Rejecting a true H‚ÇÄ | False Alarm | Healthy person labeled sick |
| **Type II Error** | Not rejecting a false H‚ÇÄ | Missed Detection | Sick person labeled healthy |



### **Easy Way to Remember**

- **Type I = False Positive = ‚ÄúCrying wolf when there is no wolf.‚Äù**  
- **Type II = False Negative = ‚ÄúFailing to see a wolf when there really is one.‚Äù**

---

**In simple words:**  
- **Type I Error = Saying something happened when it didn‚Äôt.**  
- **Type II Error = Missing something that actually happened.**

### **Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer:**



### **1. What is a Z-test?**

A **Z-test** is used when:

- The **sample size is large** (n > 30), OR  
- The **population standard deviation (œÉ) is known**

It uses the **Z-distribution**, which is the standard normal distribution (mean = 0, SD = 1).

#### **Use Z-test when:**
- You know the population standard deviation  
- You have a large sample  
- Data is approximately normal  



### **2. What is a T-test?**

A **T-test** is used when:

- The **sample size is small** (n ‚â§ 30)  
- The **population standard deviation is NOT known**  
- You must estimate the standard deviation from the sample

It uses the **t-distribution**, which is similar to the normal distribution but has **fatter tails**.

#### **Use T-test when:**
- Population standard deviation is unknown  
- Sample size is small  
- Data is approximately normal  



### **3. Key Differences (Simple Table)**

| Feature | **Z-test** | **T-test** |
|--------|------------|------------|
| Sample Size | Large (n > 30) | Small (n ‚â§ 30) |
| Population SD (œÉ) | Known | Unknown |
| Distribution Used | Z (normal) | t-distribution |
| Shape | Narrow | Wider (fat tails) |
| Best For | Large datasets | Small datasets |



### **4. Simple Example**

#### **Z-test Example:**
A factory produces 50,000 packets of chips daily.  
You want to test if the mean weight is 50g and **you know œÉ**.  
‚Üí Use Z-test.

#### **T-test Example:**
You collect a sample of 15 students‚Äô test scores and want to compare means.  
Population standard deviation is unknown.  
‚Üí Use T-test.



### **5. Easy Way to Remember**

- **Z = Known œÉ + Large sample**  
- **T = Unknown œÉ + Small sample**

Or simply:

**Z for big samples**  
 **T for small samples**



**In simple words:**  
Use **Z-test** when the data is large or you know the population‚Äôs spread.  
Use **T-test** when the data is small and you must estimate the spread from the sample.

### **Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.**

**Answer:**

---

### **1. Goal of the Question**

We want to use a **Z-test** to check whether the **mean of the sample** is equal to a **population mean of 50**.

The dataset:

```
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
```

---

### **2. Hypotheses**

- **H‚ÇÄ (Null Hypothesis):** Œº = 50  
  (The true average is 50)

- **H‚ÇÅ (Alternative Hypothesis):** Œº ‚â† 50  
  (The true average is not 50)

This is a **two-tailed test**.

---

### **3. Z-Test Formula**

\[
Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}
\]

Where:

- \( \bar{x} \) = sample mean  
- \( \mu \) = population mean (50)  
- \( \sigma \) = sample standard deviation  
- \( n \) = sample size  



### **5. Interpretation of Results (Simple Explanation)**

After running the code, you will get:

- **Sample mean** (close to 50)  
- **Z-statistic** (usually a small number like ¬±0.5)  
- **p-value**

Now compare:

- If **p-value < 0.05** ‚Üí Reject H‚ÇÄ  
- If **p-value ‚â• 0.05** ‚Üí Fail to reject H‚ÇÄ  

For this dataset, the p-value will be **greater than 0.05**, meaning:

###  **We fail to reject H‚ÇÄ.**  
There is **not enough evidence** to say the sample mean is different from 50.

So, the data suggests that the **average is still around 50**, matching the population claim.

---

### **In simple words:**

The sample data does **not** provide strong evidence that the true mean is different from 50.  
The sample average is very close to 50, so the claim seems reasonable.

---


````python
import numpy as np
from scipy import stats

# Given sample dataset
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# Step 1: Calculate sample mean
sample_mean = np.mean(sample_data)

# Step 2: Calculate sample standard deviation
sample_std = np.std(sample_data, ddof=1)

# Step 3: Sample size
n = len(sample_data)

# Step 4: Population mean (H0)
population_mean = 50

# Step 5: Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

# Step 6: Calculate p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print results with names
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Sample Size (n):", n)
print("Z-Statistic:", z_statistic)
print("P-Value:", p_value)
````

### **Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.**

**Answer:**



- A plot of 500 randomly generated normal values  
- Printed output showing:  
  - Estimated mean (close to 50)  
  - 95% confidence interval (CI range)

---



````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate normal distribution data
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)

# Calculate statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# 95% confidence interval
confidence = 0.95
z_critical = stats.norm.ppf((1 + confidence) / 2)
margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Plot the data
plt.hist(data, bins=15, color='skyblue', edgecolor='black')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')
plt.title('Simulated Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

### **Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.**

**Answer:**



### **1. Simple Explanation**

A **Z-score** tells us **how far a data point is from the mean**, measured in **standard deviations**.

- A **Z-score of 0** ‚Üí The value is exactly at the mean  
- A **Z-score of +1** ‚Üí The value is 1 standard deviation **above** the mean  
- A **Z-score of ‚Äì2** ‚Üí The value is 2 standard deviations **below** the mean  

Z-scores help us understand **how unusual or typical** a value is.



### **2. Formula for Z-score**

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:

- \( X \) = data point  
- \( \mu \) = mean of the dataset  
- \( \sigma \) = standard deviation  


### **4. Interpretation**

- After converting data to Z-scores, the **mean becomes 0**  
- The **standard deviation becomes 1**  
- The histogram shows how far values lie from the mean in units of SD  
- Helps detect:
  - Outliers  
  - Distribution shape  
  - Standardized behavior across datasets  

---

### **In simple words:**  
 **Z-scores show how many standard deviations a value is above or below the mean.**  
They help us compare different datasets on the same scale.

---


````python
### **3. Python Code to Calculate Z-scores and Plot Histogram**

import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores, mean, std

# Generate a random dataset
data = np.random.normal(100, 15, 300)

# Calculate Z-scores
z_scores, mean, std = calculate_z_scores(data)

# Print results
print("Dataset Mean:", mean)
print("Dataset Standard Deviation:", std)
print("First 10 Z-scores:", z_scores[:10])

# Plot histogram
plt.hist(z_scores, bins=20, edgecolor='black')
plt.title("Histogram of Z-score Standardized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

````
